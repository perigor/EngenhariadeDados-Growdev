Escreva uma função chamada verificar_palindromo que recebe uma string como parâmetro e retorna True se for um palíndromo (uma palavra que é igual se lida da esquerda para a direita e vice-versa) e False caso contrário.

```python
def verificar_palindromo(palavra):
    # Converte a palavra para minúsculas para garantir que a verificação não seja sensível a maiúsculas/minúsculas
    palavra = palavra.lower()
    # Verifica se a palavra é igual à sua inversa
    return palavra == palavra[::-1]

resultado1 = verificar_palindromo("arara")
resultado2 = verificar_palindromo("banana")

print("A palavra 'arara' é um palíndromo?", resultado1)
print("A palavra 'banana' é um palíndromo?", resultado2)
```

##Desafio listas e funções:

Crie uma função que representa o jogo batalha naval.

Dentro da função crie uma estrutura bidimensional utilizando listas com sublistas para representar um tabuleiro (1 lista com 20 elementos e cada elemento é uma lista de 20 elementos, tabuleiro 20x20). Cada posição irá armazenar 1 valor numérico que significa:
- 0 - Água
- 1 - Navio

Para cada posição escolha esses valores aleatoriamente, respeitando a regra de que não podem existir mais de 20 navios no tabuleiro.

Após os valores serem distribuídos, o programa deve pedir ao usuário uma posição do tabuleiro e informar se ele acertou um navio ou água e repetir o procedimento até que o usuário derrote todos os navios ou chegue ao limite de 35 tentativas.



```python
import random

def batalha_naval():
    # Inicializar o tabuleiro 20x20 com água (0)
    tabuleiro = [[0 for _ in range(20)] for _ in range(20)]

    # Colocar até 20 navios (1) no tabuleiro
    navios_restantes = 20
    while navios_restantes > 0:
        x = random.randint(0, 19)
        y = random.randint(0, 19)
        if tabuleiro[x][y] == 0:
            tabuleiro[x][y] = 1
            navios_restantes -= 1

    tentativas = 0
    navios_afundados = 0

    print("Bem-vindo ao jogo Batalha Naval!")
    print("Você tem até 10 tentativas para acertar todos os 20 navios.")

    while tentativas < 10 and navios_afundados < 20:
        # Pede ao usuário uma posição do tabuleiro
        try:
            x = int(input("Digite a coordenada X (0-19): "))
            y = int(input("Digite a coordenada Y (0-19): "))
        except ValueError:
            print("Por favor, digite números válidos.")
            continue

        if x < 0 or x > 19 or y < 0 or y > 19:
            print("Coordenadas fora do limite. Por favor, insira valores entre 0 e 19.")
            continue

        if tabuleiro[x][y] == 1:
            print("Você acertou um navio!")
            navios_afundados += 1
            tabuleiro[x][y] = 2  # Marca o navio afundado
        elif tabuleiro[x][y] == 0:
            print("Você acertou a água.")
            tabuleiro[x][y] = -1  # Marca a água já acertada
        else:
            print("Você já tentou essa posição. Tente outra.")
            continue

        tentativas += 1
        print(f"Tentativas restantes: {10 - tentativas}")
        print(f"Navios afundados: {navios_afundados}/20")

    if navios_afundados == 20:
        print("Parabéns! Você afundou todos os navios!")
    else:
        print("Fim do jogo. Você não conseguiu afundar todos os navios.")

# Chama a função para iniciar o jogo
batalha_naval()10
```

```python
import csv

def ler_csv_para_dicionario(caminho_arquivo):
    dados = {}
    with open(caminho_arquivo, newline='', encoding='utf-8') as csvfile:
        leitor = csv.reader(csvfile)
        cabecalho = next(leitor)  # Lê a linha de cabeçalho

        for i, linha in enumerate(leitor):
            individuo = {cabecalho[j]: linha[j] for j in range(len(cabecalho))}
            dados[i] = individuo

    return dados
```

Qual a porcentagem de homens e mulheres na base de dados?

```python
import csv

def calcular_porcentagem_genero(caminho_arquivo):
    total_homens = 0
    total_mulheres = 0
    total_registros = 0

    # Abre o arquivo CSV para leitura
    with open(caminho_arquivo, newline='', encoding='utf-8') as csvfile:
        leitor_csv = csv.reader(csvfile)
        # Ignora o cabeçalho
        next(leitor_csv)

        # Itera sobre as linhas do arquivo CSV
        for linha in leitor_csv:
            # Supondo que o gênero esteja na quinta coluna (índice 4) de cada linha
            genero = linha[3].upper()  # Converte para maiúsculas para evitar discrepâncias de capitalização
            if genero == 'M':
                total_homens += 1
            elif genero == 'F':
                total_mulheres += 1
            total_registros += 1

    # Calcula a porcentagem de homens e mulheres
    porcentagem_homens = (total_homens / total_registros) * 100
    porcentagem_mulheres = (total_mulheres / total_registros) * 100

    return porcentagem_homens, porcentagem_mulheres

# Caminho do arquivo CSV
caminho_arquivo = '/content/compras.csv'

# Chama a função para calcular a porcentagem de homens e mulheres na base de dados
porcentagem_homens, porcentagem_mulheres = calcular_porcentagem_genero(caminho_arquivo)

# Imprime os resultados
print("Porcentagem de homens na base de dados:", porcentagem_homens)
print("Porcentagem de mulheres na base de dados:", porcentagem_mulheres)
```

Qual foi o gasto por ano?

```python
import csv

def calcular_gasto_por_ano(caminho_arquivo):
    gasto_por_ano = {}

    # Abre o arquivo CSV para leitura
    with open(caminho_arquivo, newline='', encoding='utf-8') as csvfile:
        leitor_csv = csv.reader(csvfile)
        # Ignora o cabeçalho
        next(leitor_csv)

        # Itera sobre as linhas do arquivo CSV
        for linha in leitor_csv:
            # Supondo que o ano esteja na quinta coluna (índice 4) de cada linha
            ano = linha[5]
            # Supondo que o valor da compra esteja na última coluna de cada linha
            valor_compra = float(linha[4])

            # Atualiza o total de gastos para o ano correspondente
            if ano in gasto_por_ano:
                gasto_por_ano[ano] += valor_compra
            else:
                gasto_por_ano[ano] = valor_compra

    return gasto_por_ano

# Caminho do arquivo CSV
caminho_arquivo = '/content/compras.csv'

# Chama a função para calcular o gasto por ano na base de dados
gasto_por_ano = calcular_gasto_por_ano(caminho_arquivo)

# Imprime os resultados
print("Gasto por ano na base de dados:")
for ano, gasto in gasto_por_ano.items():
    print(f"Ano {ano}: R$ {gasto:.2f}")
```

Utilizando as faixas etárias, diga quantas pessoas há em cada faixa?

```python
import csv

def contar_pessoas_por_faixa_etaria(caminho_arquivo):
    pessoas_por_faixa_etaria = {
        'Jovens': 0,
        'Adultos': 0,
        'Idosos': 0
    }

    # Abre o arquivo CSV para leitura
    with open(caminho_arquivo, newline='', encoding='utf-8') as csvfile:
        leitor_csv = csv.reader(csvfile)
        # Ignora o cabeçalho
        next(leitor_csv)

        # Itera sobre as linhas do arquivo CSV
        for linha in leitor_csv:
            idade = int(linha[2])  # Obtém a idade da pessoa da terceira coluna (índice 2)
            if 18 <= idade <= 25:
                pessoas_por_faixa_etaria['Jovens'] += 1
            elif 26 <= idade <= 59:
                pessoas_por_faixa_etaria['Adultos'] += 1
            else:
                pessoas_por_faixa_etaria['Idosos'] += 1

    return pessoas_por_faixa_etaria

# Caminho do arquivo CSV
caminho_arquivo = '/content/compras.csv'

# Chama a função para contar pessoas por faixa etária na base de dados
pessoas_por_faixa_etaria = contar_pessoas_por_faixa_etaria(caminho_arquivo)

# Imprime os resultados
print("Número de pessoas em cada faixa etária:")
for faixa_etaria, quantidade in pessoas_por_faixa_etaria.items():
    print(f"{faixa_etaria}: {quantidade}")
```

Qual é a faixa etária que mais gasta?

```python
import csv

def calcular_gasto_por_faixa_etaria(caminho_arquivo):
    gasto_por_faixa_etaria = {'Jovens': 0, 'Adultos': 0, 'Idosos': 0}

    # Abre o arquivo CSV para leitura
    with open(caminho_arquivo, newline='', encoding='utf-8') as csvfile:
        leitor_csv = csv.reader(csvfile)
        # Ignora o cabeçalho
        next(leitor_csv)

        # Itera sobre as linhas do arquivo CSV
        for linha in leitor_csv:
            idade = int(linha[2])  # Obtém a idade da pessoa da terceira coluna (índice 2)
            valor_compra = float(linha[4])  # Obtém o valor da compra da quinta coluna (índice 4)

            # Determina a faixa etária da pessoa
            if 18 <= idade <= 25:
                faixa_etaria = 'Jovens'
            elif 26 <= idade <= 59:
                faixa_etaria = 'Adultos'
            else:
                faixa_etaria = 'Idosos'

            # Adiciona o valor da compra à faixa etária correspondente
            gasto_por_faixa_etaria[faixa_etaria] += valor_compra

    return gasto_por_faixa_etaria

# Caminho do arquivo CSV
caminho_arquivo = '/content/compras.csv'

# Chama a função para calcular o gasto por faixa etária
gasto_por_faixa_etaria = calcular_gasto_por_faixa_etaria(caminho_arquivo)

# Identifica a faixa etária que mais gastou
faixa_etaria_mais_gastou = max(gasto_por_faixa_etaria, key=gasto_por_faixa_etaria.get)

# Imprime o resultado
print("Faixa etária que mais gastou:", faixa_etaria_mais_gastou)
```

A opção de débito é mais utilizada entre homens ou mulheres?

```python
import csv

def identificar_uso_debito_por_genero(caminho_arquivo):
    homens_debito = 0
    mulheres_debito = 0

    # Abre o arquivo CSV para leitura
    with open(caminho_arquivo, newline='', encoding='utf-8') as csvfile:
        leitor_csv = csv.reader(csvfile)
        # Ignora o cabeçalho
        next(leitor_csv)

        # Itera sobre as linhas do arquivo CSV
        for linha in leitor_csv:
            genero = linha[3]  # Obtém o gênero da pessoa da quarta coluna (índice 3)
            tipo_compra = linha[6].lower()  # Obtém o tipo de compra da sétima coluna (índice 6) e converte para minúsculas
            if tipo_compra == 'debito':
                if genero == 'M':
                    homens_debito += 1
                elif genero == 'F':
                    mulheres_debito += 1

    return homens_debito, mulheres_debito

# Caminho do arquivo CSV
caminho_arquivo = '/content/compras.csv'

# Chama a função para identificar o uso de débito por gênero na base de dados
homens_debito, mulheres_debito = identificar_uso_debito_por_genero(caminho_arquivo)

# Imprime os resultados
print("Número de compras no débito por gênero na base de dados:")
print("Homens:", homens_debito)
print("Mulheres:", mulheres_debito)

# Comparação para determinar se o débito é mais utilizado entre homens ou mulheres
if homens_debito > mulheres_debito:
    print("O débito é mais utilizado entre homens.")
elif homens_debito < mulheres_debito:
    print("O débito é mais utilizado entre mulheres.")
else:
    print("O débito é igualmente utilizado entre homens e mulheres.")
```

Qual o sobrenome que mais aparece na base de dados?

```python
import csv
from collections import Counter

def encontrar_sobrenome_mais_frequente(caminho_arquivo):
    sobrenomes = []

    # Abre o arquivo CSV para leitura
    with open(caminho_arquivo, newline='', encoding='utf-8') as csvfile:
        leitor_csv = csv.reader(csvfile)
        # Ignora o cabeçalho
        next(leitor_csv)

        # Itera sobre as linhas do arquivo CSV
        for linha in leitor_csv:
            sobrenome = linha[1].split()[-1]  # Obtém o último sobrenome da segunda coluna (índice 1)
            sobrenomes.append(sobrenome)

    # Conta a frequência de cada sobrenome
    contador_sobrenomes = Counter(sobrenomes)
    sobrenome_mais_frequente = contador_sobrenomes.most_common(1)[0][0]

    return sobrenome_mais_frequente

# Caminho do arquivo CSV
caminho_arquivo = '/content/compras.csv'

# Chama a função para encontrar o sobrenome mais frequente na base de dados
sobrenome_mais_frequente = encontrar_sobrenome_mais_frequente(caminho_arquivo)

# Imprime o resultado
print("Sobrenome mais frequente na base de dados:", sobrenome_mais_frequente)
```

Qual foi o ano em que os homens mais usaram o crédito?

```python
import csv

def encontrar_ano_homem_mais_usou_credito(caminho_arquivo):
    ano_mais_uso_credito = None
    max_compras_credito = 0

    # Dicionário para armazenar o número de compras com crédito por ano
    compras_credito_por_ano = {}

    # Abre o arquivo CSV para leitura
    with open(caminho_arquivo, newline='', encoding='utf-8') as csvfile:
        leitor_csv = csv.reader(csvfile)
        # Ignora o cabeçalho
        next(leitor_csv)

        # Itera sobre as linhas do arquivo CSV
        for linha in leitor_csv:
            genero = linha[3]  # Obtém o gênero da pessoa da quarta coluna (índice 3)
            tipo_compra = linha[6].lower()  # Obtém o tipo de compra da sétima coluna (índice 6) e converte para minúsculas
            ano = linha[5]  # Obtém o ano da compra da sexta coluna (índice 5)

            # Verifica se a compra foi feita por um homem e usando cartão de crédito
            if genero == 'M' and tipo_compra == 'credito':
                # Atualiza o número de compras com crédito para o ano correspondente
                compras_credito_por_ano[ano] = compras_credito_por_ano.get(ano, 0) + 1

                # Verifica se o número de compras com crédito para este ano é o maior até agora
                if compras_credito_por_ano[ano] > max_compras_credito:
                    max_compras_credito = compras_credito_por_ano[ano]
                    ano_mais_uso_credito = ano

    return ano_mais_uso_credito

# Caminho do arquivo CSV
caminho_arquivo = '/content/compras.csv'

# Chama a função para encontrar o ano em que os homens mais usaram o crédito na base de dados
ano_mais_uso_credito = encontrar_ano_homem_mais_usou_credito(caminho_arquivo)

# Imprime o resultado
print("Ano em que os homens mais usaram o crédito:", ano_mais_uso_credito)
```

Qual opção de pagamento é mais utilizada em cada faixa etária?

```python
import csv

def opcao_pagamento_por_faixa_etaria(caminho_arquivo):
    opcao_pagamento_por_faixa = {
        'Jovens': {'debito': 0, 'credito': 0, 'dinheiro': 0},
        'Adultos': {'debito': 0, 'credito': 0, 'dinheiro': 0},
        'Idosos': {'debito': 0, 'credito': 0, 'dinheiro': 0}
    }

    # Abre o arquivo CSV para leitura
    with open(caminho_arquivo, newline='', encoding='utf-8') as csvfile:
        leitor_csv = csv.reader(csvfile)
        # Ignora o cabeçalho
        next(leitor_csv)

        # Itera sobre as linhas do arquivo CSV
        for linha in leitor_csv:
            idade = int(linha[2])  # Obtém a idade da pessoa da terceira coluna (índice 2)
            tipo_compra = linha[6].lower()  # Obtém o tipo de compra da sétima coluna (índice 6) e converte para minúsculas

            # Determina a faixa etária da pessoa
            if 18 <= idade <= 25:
                faixa_etaria = 'Jovens'
            elif 26 <= idade <= 59:
                faixa_etaria = 'Adultos'
            else:
                faixa_etaria = 'Idosos'

            # Incrementa o contador correspondente à faixa etária e ao tipo de pagamento
            opcao_pagamento_por_faixa[faixa_etaria][tipo_compra] += 1

    return opcao_pagamento_por_faixa

# Caminho do arquivo CSV
caminho_arquivo = '/content/compras.csv'

# Chama a função para contar a opção de pagamento por faixa etária na base de dados
opcao_pagamento_por_faixa = opcao_pagamento_por_faixa_etaria(caminho_arquivo)

# Imprime os resultados
print("Opção de pagamento por faixa etária:")
for faixa_etaria, opcoes in opcao_pagamento_por_faixa.items():
    print(f"Faixa etária: {faixa_etaria}")
    for opcao, quantidade in opcoes.items():
        print(f"{opcao.capitalize()}: {quantidade}")
    print()
```

Qual o valor gasto em compras por jovens do ano de 2010 até 2015?

```python
import csv

def calcular_gasto_jovens(caminho_arquivo):
    gasto_jovens = 0

    # Abre o arquivo CSV para leitura
    with open(caminho_arquivo, newline='', encoding='utf-8') as csvfile:
        leitor_csv = csv.reader(csvfile)
        # Ignora o cabeçalho
        next(leitor_csv)

        # Itera sobre as linhas do arquivo CSV
        for linha in leitor_csv:
            idade = int(linha[2])  # Obtém a idade da pessoa da terceira coluna (índice 2)
            ano_compra = int(linha[5])  # Obtém o ano da compra da sexta coluna (índice 5)

            # Verifica se a pessoa é jovem (idade entre 18 e 25 anos) e se a compra ocorreu entre 2010 e 2015
            if 18 <= idade <= 25 and 2010 <= ano_compra <= 2015:
                valor_compra = float(linha[4])  # Obtém o valor da compra da quinta coluna (índice 4)
                gasto_jovens += valor_compra

    return gasto_jovens

# Caminho do arquivo CSV
caminho_arquivo = '/content/compras.csv'

# Chama a função para calcular o valor gasto por jovens no período de 2010 até 2015
gasto_jovens = calcular_gasto_jovens(caminho_arquivo)

# Imprime o resultado
print("Valor gasto em compras por jovens do ano de 2010 até 2015:", gasto_jovens)
```

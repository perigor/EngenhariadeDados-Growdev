```python
# Importando biblioteca

import pandas as pd
```

```python
# dados

df_pnad = pd.read_csv("/content/dados.csv")
df_pnad.head()
```

Qual a quantidade de pessoas nas seguintes faixas etárias
- 20 anos ou menos
- 21 a 35 anos
- 36 a 50 anos
- 51 a 65 anos
- maiores de 65 anos



```python

# Definindo as faixas etárias
faixas_etarias = [
    (df_pnad['Idade'] <= 20, '20 anos ou menos'),
    ((df_pnad['Idade'] >= 21) & (df_pnad['Idade'] <= 35), '21 a 35 anos'),
    ((df_pnad['Idade'] >= 36) & (df_pnad['Idade'] <= 50), '36 a 50 anos'),
    ((df_pnad['Idade'] >= 51) & (df_pnad['Idade'] <= 65), '51 a 65 anos'),
    (df_pnad['Idade'] > 65, 'maiores de 65 anos')
]

# Contagem
contagem_faixas = {descricao: len(df_pnad[condicao]) for condicao, descricao in faixas_etarias}

# Prints
for faixa, quantidade in contagem_faixas.items():
    print(f'{faixa}: {quantidade} pessoas')
```

Qual é a média de escolaridade em cada faixa etária?


```python
# Definir as faixas etárias
bins = [0, 20, 35, 50, 65, 100]  #Intervalo com bas na faixa etária
labels = ['0 a 20 anos', '21 a 35 anos', '36 a 50 anos', '51 a 65 anos', 'maiores de 65 anos'] #Fornece rótulos das faixas etárias

# Nova coluna para as faixas etárias
df_pnad['Faixa Etária'] = pd.cut(df_pnad['Idade'], bins=bins, labels=labels, right=True)

# Média de anos de estudo em cada faixa etária
media_escolaridade_faixa_etaria = df_pnad.groupby('Faixa Etária')['Anos de Estudo'].mean()

# Print
media_escolaridade_faixa_etaria.reset_index()
```

Qual a UF que concentra a maior renda?


```python
# Localizando a linha com valor máximo dentro de RENDA
indice_max_renda = df_pnad['Renda'].idxmax()

# Localizando  a UF associada ao valor máximo
uf_maior_renda = df_pnad.loc[indice_max_renda, 'UF']
renda_maxima = df_pnad.loc[indice_max_renda, 'Renda']

print(f"A UF que concentra a maior renda é a Bahia, número {uf_maior_renda} com uma renda de R${renda_maxima:.2f}")

```

```python
# Identificação de cor conforme DF
cor = {0: 'Indígena', 2: 'Branca', 4: 'Preta', 6: 'Amarela', 8: 'Parda', 9: 'Sem declaração'}

# Identificação de sexo conforme DF
sexo = {0: 'Masculino', 1: 'Feminino'}

# Criando 'Cor Descrição' e 'Sexo Descrição' usando os mapeamentos
df_pnad['Cor Descrição'] = df_pnad['Cor'].map(cor)
df_pnad['Sexo Descrição'] = df_pnad['Sexo'].map(sexo)

# Filtrar apenas as pessoas pretas e brancas
df_brancos_pretos = df_pnad[df_pnad['Cor'].isin([2, 4])]

# Agrupando por Sexo Descrição e Cor Descrição e calculando a renda média
renda_media = df_brancos_pretos.groupby(['Sexo Descrição', 'Cor Descrição'])['Renda'].mean()

# Print
renda_media.reset_index()
```

Qual a renda média concentrada em cada faixa etária?




```python
# Definir faixas etárias
bins = [0, 20, 35, 50, 65, 100]
labels = ['0 a 20 anos', '21 a 35 anos', '36 a 50 anos', '51 a 65 anos', 'maiores de 65 anos']

# 2. Criando uma nova coluna para as faixas etárias
df_pnad['Faixa Etária'] = pd.cut(df_pnad['Idade'], bins=bins, labels=labels, right=True)

# 3. Agrupar por Faixa Etária e calcular a renda média
renda_media_faixa_etaria = df_pnad.groupby('Faixa Etária')['Renda'].mean()

# Mostrar o resultado
renda_media_faixa_etaria.reset_index()
```

Qual a escolaridade entre indígenas, para homens e mulheres?


```python
# Adicionando colunas 'Cor Descrição' e 'Sexo Descrição' usando os mapeamentos
df_pnad['Cor Descrição'] = df_pnad['Cor'].map(cor)
df_pnad['Sexo Descrição'] = df_pnad['Sexo'].map(sexo)

# Filtrando apenas os indígenas
df_indigenas = df_pnad[df_pnad['Cor'] == 0]

# Calculando a escolaridade média por sexo entre os indígenas
escolaridade_media_indigenas = df_indigenas.groupby('Sexo Descrição')['Anos de Estudo'].mean()

# Print
escolaridade_media_indigenas.reset_index()
```

Qual a média de anos de estudo entre brancos e negros?


```python


# Adicionar coluna 'Cor Descrição' usando o mapeamento
df_pnad['Cor Descrição'] = df_pnad['Cor'].map(cor)

# Filtrando brancos e negros
df_brancos_negros = df_pnad[df_pnad['Cor'].isin([2, 4])]

# Calculando a média de anos de estudo para cada grupo
media_anos_estudo_brancos_negros = df_brancos_negros.groupby('Cor Descrição')['Anos de Estudo'].mean()

# Print
media_anos_estudo_brancos_negros.reset_index()
```

Qual a média de anos de estudo e de renda para mulheres brancas, e para mulheres negras?

```python
# Filtrando mulheres brancas
mulheres_brancas = df_pnad[(df_pnad['Sexo'] == 1) & (df_pnad['Cor'] == 2)]

# Filtrando  mulheres negras
mulheres_negras = df_pnad[(df_pnad['Sexo'] == 1) & (df_pnad['Cor'] == 4)]

# Calcullando média de anos de estudo e renda para cada grupo
media_mulheres_brancas = mulheres_brancas[['Anos de Estudo', 'Renda']].mean()
media_mulheres_negras = mulheres_negras[['Anos de Estudo', 'Renda']].mean()

# Print
print("Média de anos de estudo e renda para mulheres brancas:")
print(media_mulheres_brancas)

print("\nMédia de anos de estudo e renda para mulheres negras:")
print(media_mulheres_negras)
```

 Qual é a UF com maior média de escolaridade e qual a UF com maior média de renda?

```python
# Média de escolaridade por UF
media_escolaridade_uf = df_pnad.groupby('UF')['Anos de Estudo'].mean()

# Média de renda por UF
media_renda_uf = df_pnad.groupby('UF')['Renda'].mean()

# 3. Identificando a UF com maior média de escolaridade e maior média de renda
uf_maior_media_escolaridade = media_escolaridade_uf.idxmax()
uf_maior_media_renda = media_renda_uf.idxmax()

# Mostrar os resultados
print(f"UF com maior média de escolaridade: {uf_maior_media_escolaridade} ({media_escolaridade_uf.max():.2f} anos de estudo)")
print(f"UF com maior média de renda: {uf_maior_media_renda} (R$ {media_renda_uf.max():.2f})")

#53	Distrito Federal
```

```
import great_expectations as ge
from pyspark.sql import functions as F
```

Carregamento

```
tmp_parquet_path = "/tmp/transient/tb_Products"
df = spark.read.format("parquet").load(tmp_parquet_path)
```

```
display(df)
```

Verificações

```
df_ge = ge.dataset.SparkDFDataset(df)

#Verificar se a coluna 'NETAMOUNT' não possui valores nulos
expectativa_nulos = df_ge.expect_column_values_to_not_be_null(column='PRICE')

#Outra coluna para verificação
expectativa_nulos_outra = df_ge.expect_column_values_to_not_be_null(column='PRODUCTID')

#Validar as expectativas
resultado_validacao = df_ge.validate()

if resultado_validacao['success']:
    print('Arquivo Ok!')
else:
    print('Erro na validação!')
```

```
df = df.withColumn("dt_ingestion", F.date_format(F.current_date(), "dd-MM-yyyy"))
```

```
display(df)
```

Salvar no formato Delta

```
delta = "/tmp/bronze/tb_Products"
df.write.partitionBy("dt_ingestion").format("delta").mode("overwrite").save(delta)
```
